```python
from dotenv import load_dotenv
import os
import spotipy
from spotipy.oauth2 import SpotifyOAuth
#from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
#Se carga las credenciales del entorno
load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id = os.getenv('CLIENT_ID'),
    client_secret = os.getenv('CLIENT_SECRET'),
    redirect_uri = 'http://127.0.0.1:8888/callback', #seria ideal poner un localhost para las pruebas pero se niega a colaborar
    scope='user-top-read user-library-read playlist-read-private',
    open_browser=True,
    show_dialog=True
))
```

```python
response_top_artist = sp.current_user_top_artists(limit=50, time_range='long_term')
print(response_top_artist)#Observar la forma de la respuesta
```

```python
#Mis artistas mas escuchados
for id, artist in enumerate(response_top_artist['items']):
    print(f"{id+1}. {artist['name']} ; {artist['popularity']}")
```

```python
response_top_tracks = sp.current_user_top_tracks(limit=50, time_range='long_term')
print(response_top_tracks)
```

```python
#Generar el dataframe

tracks = response_top_tracks['items']
albums = []

for t in tracks:
    albums.append({
        "track_id": t['id'],
        "track_name": t['name'],
        "artist": t['artists'][0]['name'],
        "album_name": t['album']['name'],
        "album_realise_date": t['album']['release_date'],
        "popularity": t['popularity']
    })

df_tracks = pd.DataFrame(albums)
df_tracks.head()




```

```python
# Crear columna de día de la semana (0=Lunes, 6=Domingo)
df_songs_history_clean['day_of_week'] = df_songs_history_clean['ts'].dt.day_name()

# Agrupar minutos escuchados por día y hora
heatmap_data = (
    df_songs_history_clean.groupby(['day_of_week', 'hour'])['minutes_played']
    .sum()
    .reset_index()
)

# Reordenar días de la semana
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
heatmap_data['day_of_week'] = pd.Categorical(heatmap_data['day_of_week'], categories=days_order, ordered=True)

# Crear tabla pivote
heatmap_pivot = heatmap_data.pivot(index="day_of_week", columns="hour", values="minutes_played").fillna(0)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
sns.heatmap(heatmap_pivot, cmap="YlOrBr", linewidths=0.5)

plt.title("🔥 Minutos de música escuchados (día de la semana × hora)", fontsize=14, pad=15)
plt.xlabel("Hora del día")
plt.ylabel("Día de la semana")
plt.show()

```

```python
# Normalizar: convertir minutos en % dentro de cada día
heatmap_pivot_pct = heatmap_pivot.div(heatmap_pivot.sum(axis=1), axis=0) * 100

plt.figure(figsize=(14,6))
sns.heatmap(heatmap_pivot_pct, cmap="YlGnBu", linewidths=0.5, annot=False)

plt.title("📊 Distribución porcentual de música escuchada (día × hora)", fontsize=14, pad=15)
plt.xlabel("Hora del día")
plt.ylabel("Día de la semana")
plt.show()
```

```python
import time

# Diccionario para cachear artistas ya consultados
artist_genres_dict = {}

# Lista donde guardaremos resultados
track_genres = []

for track_id in df_songs_history_clean['track_id'].unique():
    try:
        # Info del track
        track_info = sp.track(track_id)
        artist_id = track_info['artists'][0]['id']
        artist_name = track_info['artists'][0]['name']

        # Evitar repetir llamadas si ya tenemos el artista
        if artist_id not in artist_genres_dict:
            artist_info = sp.artist(artist_id)
            genres = artist_info.get('genres', [])
            artist_genres_dict[artist_id] = genres
            time.sleep(0.1)  # evitar rate limit
        else:
            genres = artist_genres_dict[artist_id]

        # Guardar resultado
        track_genres.append({
            "track_id": track_id,
            "artist": artist_name,
            "genres": genres
        })

    except Exception as e:
        print(f"Error con {track_id}: {e}")
```

```python
df_genres = pd.DataFrame(track_genres).dropna()

df_genres.info()
```
